```python
#import essential libaries
import yfinance as yf
import pandas as pd
```

Using the yfinance Library to Extract Stock Data(Apple,Amd)

```python
#ticker module creates an object that helps to work on extraction of data from https://pypi.org/user/ranaroussi/
apple = yf.Ticker("AAPL")
```

Using the attribute  <code>info</code> we can extract information about the stock as a Python dictionary.


```python
#to get information about apple on yfinance
apple_info=apple.info
apple_info
```

```python
#country info
apple_info['country']
```

A share is the single smallest part of a company's stock  that you can buy, the prices of these shares fluctuate over time. Using the <code>history()</code> method we can get the share price of the stock over a certain period of time. Using the `period` paramete
r we can set how far back from the present to get data. The options for `period` are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.


```python
#histy of rise and clsoe and open stock market prices
apple_share_price_data = apple.history(period="max")
apple_share_price_data
```

The format that the data is returned in is a Pandas DataFrame. With the `Date` as the index the share `Open`, `High`, `Low`, `Close`, `Volume`, and `Stock Splits` are given for each day.


```python
#top of data
apple_share_price_data.head()
```

We can reset the index of the DataFrame with the `reset_index` function. We also set the `inplace` paramter to `True` so the change takes place to the DataFrame itself.

```python
apple_share_price_data.reset_index(inplace=True)
```

```python
apple_share_price_data.plot(x="Date", y="Open")
```
